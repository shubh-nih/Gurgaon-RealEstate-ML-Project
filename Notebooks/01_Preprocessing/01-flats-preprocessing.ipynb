```python
# columns to drop -> links, property_id
flats.drop(columns = ['link', 'property_id'], inplace = True)
```

```python
flats.sample(2)
```

```python
# rename column
flats.rename(columns = {'area':'price_per_sqft'}, inplace = True)
```

```python
flats.sample(2)
```

```python
# copy of dataset
df = flats.copy()
```

```python
# society 
df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r"\d+(\.\d+)?\s*ï¿½", "", str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df['society'].value_counts()
```

```python
# price
df['price'].value_counts()
```

```python
df[df['price'] == 'Price on Request']
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def price_change(price):
    if type(price) == float:
        return price
    else:
        if price[1] == 'Lac':
            return round(float(price[0])/100, 2)
        else:
            return round(float(price[0]), 2)
```

```python
df['price'] = df['price'].str.split(' ').apply(price_change)
```

```python
df.sample(5)
```

```python
# price_per_sqft
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('ýýý', ' ').str.replace(',', '').str.strip().astype(float)
```

```python
df.sample(4)
```

```python
# areaWithType
df['areaWithType'].value_counts()
```

```python
df[df['bedRoom'].isnull()]
```

```python
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype(int)
```

```python
df['bedRoom'].value_counts()
```

```python
# bathroom
df['bathroom'].value_counts()
```

```python
df[df['bathroom'].isnull()]
```

```python
df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype(int)
```

```python
df['bathroom'].value_counts()
```

```python
# balcony
df['balcony'].value_counts()
```

```python
df['floorNum'] = df['floorNum'].apply(lambda floor: re.sub(r"[ï¿½]", "", str(floor))).str.replace('Ground', '0').str.replace('Basement', '-1').str.split(' ').str.get(0).str.replace('Lower', '0').str.extract(r'(\d+)')
```

```python
df.sample(4)
```

```python
# feature engineering area column
df.insert(loc = 4, column = 'area', value = round((df['price']*10000000)/df['price_per_sqft']))
```
