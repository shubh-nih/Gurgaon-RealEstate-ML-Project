```python
import pandas as pd
import numpy as np
```

```python
flats = pd.read_csv('flats-cleaned.csv')
houses = pd.read_csv('houses-cleaned.csv')
```

```python
data = pd.concat([flats, houses], ignore_index = True)
```

```python
data = data.sample(data.shape[0], ignore_index = True)
```
